```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
len(data.columns)
```

```python
#separating dependent and independent features
x = data.iloc[:,3:13]
y = data.iloc[:,13:]
```

```python
#converting geography and gender to categorical
geo = pd.get_dummies(x["Geography"],drop_first=True)
```

```python
gender = pd.get_dummies(x["Gender"],drop_first=True)
```

```python
#concating
x = pd.concat([x,geo,gender],axis=1)
```

```python
x = x.drop(["Geography","Gender"],axis=1)
```

```python
x.head(3)
```

```python
  #splitting the data into train and test data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.head(3)
```

```python
## feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x_train = ss.fit_transform(x_train)
```

```python
x_test = ss.fit_transform(x_test)
```

```python
x_train #-1 t0
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#initializing ANN
#empty ANN
classifier = Sequential()
```

```python
#adding 1st hidden layer
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu',input_dim =11))
# 11 -- col
```

```python
#adding 2nd hidden layer
classifier.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
```

```python
classifier.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
```

```python
#adding output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
#compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fitting
model = classifier.fit(x_train,y_train,validation_split= 0.33,batch_size=10,epochs=100)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred = y_pred>0.5
```

```python
y_pred
```

```python
#conufusion metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```
